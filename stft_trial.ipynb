```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import IPython.display as ipd
import librosa, librosa.display
```

## Loading file

```python
x, sr = librosa.load('Sample1HQ.wav')
ipd.Audio(x, rate=sr)
```

```python
plt.figure(figsize=(12, 4))
librosa.display.waveplot(x, sr=sr)
```

```python
X = librosa.stft(x)
S = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(15, 5))
librosa.display.specshow(S, sr=sr, x_axis='time', y_axis='linear')
plt.colorbar(format='%+2.0f dB')
```

```python
X = librosa.stft(x)
S = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(15, 5))
librosa.display.specshow(S, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
```

```python
fmin = librosa.midi_to_hz(60)
C = librosa.cqt(x, sr=sr, fmin=fmin, n_bins=22)
logC = librosa.amplitude_to_db(abs(C))
```

```python
plt.figure(figsize=(15, 5))
librosa.display.specshow(logC, sr=sr, x_axis='time', y_axis='cqt_note', fmin=fmin)
plt.colorbar(format='%+2.0f dB')
```
